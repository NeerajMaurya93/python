```python
import  matplotlib.pyplot as plt
Category=["Apple","Banana","Guava","Grapes"]
Values=[10,20,30,40]
plt.bar(Category,Values)
plt.title("Bar Chart")
plt.show()
```

```python
import  matplotlib.pyplot as plt
Category=["Apple","Banana","Guava","Grapes"]
Values=[10,20,30,40]
plt.barh(Category,Values,color="purple")
plt.title("Bar Chart")
plt.show()
```

```python
import  matplotlib.pyplot as plt
plt.style.use('dark_background')
Category=["Apple","Banana","Guava","Grapes"]
Values=[10,20,30,40]
plt.bar(Category,Values,color=['aqua','green','purple','red'],edgecolor="White")
plt.title("Bar Chart")
plt.show()
```

```python
import  matplotlib.pyplot as plt
plt.style.use('dark_background')
Category=["Apple","Banana","Guava","Grapes"]
Values=[10,20,30,40]
bars=plt.bar(Category,Values,color=['aqua','green','purple','red'],edgecolor="White")
for bar in bars:
    height =bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        height+1,
        str(height),
        ha='center',
        color='white',
        fontsize=12
    )
plt.show()
```
