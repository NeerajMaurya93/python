```python
import pandas as pd 
df=pd.read_excel(r"C:\Users\DELL\OneDrive\Documents\Unclean_MBA_Data.xlsx")
print(df.head())
```

```python
# To Show Row and Column
print(df.shape)
```

```python
# To Show Data Type, Memory uses, messing value idea
print(df.info())
```

```python
# To Print the Statistics of the numarical columns
print(df.describe()) 
```

```python
df.columns=df.columns.str.strip() # Remove white space from both sides
df.columns=df.columns.str.lower()# Convert the column names into  lower case

df.columns # To display all the column names 
```

```python
import pandas as pd 
import numpy as np 
df["units sold"] = df["units sold"].replace("Uknown",np.nan)
df["units sold"] = pd.to_numeric(df["units sold"],errors="coerce")
df["units sold"] = df["units sold"].fillna(df["units sold"].median())
print(df)
```

```python
df["mba"]=df["mba"].replace({"true":True,"True":True,"fals":False,"False":False})
print(df)
```

```python
import pandas as pd
df["total price"] = df["units sold"] * df["unit price"]
print(df)
```

```python
import pandas as pd
df["remarks"]=df["remarks"].fillna("No Remarks")
print(df)
```

```python
# To drop column
df= df.drop("unique vendors",axis=1)
print(df)
```

```python
# To Drop the missing row
df=df.dropna(subset=["item name","unit price"])
print(df)
```

```python
# To Convert The Column Title
df.columns=df.columns.str.title()
print(df.columns)
```
