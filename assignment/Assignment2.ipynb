```python
impty1=[]
impty2=[]
impty3=[]

for i in range (1,51):
    impty1.append(i*3)
    impty2.append(i*5)
    impty3.append(i*3)
    impty3.append(i*5)
print("Multiple of 3 :\n",impty1)
print("Multiple of 5 :\n",impty2)
print("Multiple of both :\n",impty3)
```

```python
sen="Silly Fox"
vow="aeiouAEIOU"
vow_count=0
c_count=0
for ch in sen:
    if ch.isalpha():
        if ch in vow:
            vow_count+=1
        else:
            c_count+=1
print("Vovels :",vow_count)
print("Consonants: ",c_count)            
```
