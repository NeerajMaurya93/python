## Key Features of NumPy

### 1.Efficient Array Operations:
NumPy arrays are stored in contiguous memory locations, making them faster and more efficient than Python lists. This is due to the locality of reference, which allows for quicker access and manipulation of data.
### 2.Multidimensional Arrays:
NumPy supports arrays with multiple dimensions, allowing for complex data structures. The number of dimensions is called the rank, and the shape of the array is a tuple of integers representing the size along each dimension.
### 3.Mathematical Functions: 
NumPy provides a wide range of mathematical functions for operations such as addition, subtraction, multiplication, division, and more. These operations can be performed element-wise on arrays.
### 4.Broadcasting: 
NumPy supports broadcasting, which allows for operations between arrays of different shapes. This feature is useful for performing operations between a vector and a scalar or between arrays of different sizes.
### 5.Indexing and Slicing: 
NumPy arrays can be indexed and sliced in various ways, similar to Python lists. This allows for easy access and manipulation of specific elements or subarrays.

# Why is NumPy Faster than Lists ?

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate then very efficiently.
This behavior is called locality of reference in computer science.
### NumPy arrays are faster than Python lists due to several key reasons:
### 1.Homogeneous Data Types:
NumPy arrays store elements of the same data type, which allows for optimized memory usage and faster operations. Python lists, on the other hand, can store mixed data types, which adds overhead.

### 2.Contiguous Memory Allocation:
NumPy arrays are stored in contiguous memory blocks, enabling efficient access and manipulation. Python lists are stored as pointers to objects, which increases memory fragmentation and slows down operations.

### 3.Vectorization:
NumPy supports vectorized operations, meaning you can perform element-wise computations without explicit loops. This leverages low-level C implementations, making operations significantly faster compared to Python's interpreted loops.

### 4.Efficient Memory Management:
NumPy uses less memory compared to Python lists because it avoids the overhead of storing type information for each element. This compact storage improves performance, especially for large datasets.

### 5.Built-in Optimizations:
NumPy is implemented in C and Fortran, which are highly optimized for numerical computations. Python lists rely on Python's general-purpose implementation, which is slower for numerical tasks.

## Array Reshaping

```python
#1D array
arr=np.arange(12)
print("1..Original array :\n",arr)

# in to 2D array
reshaped=arr.reshape((3,4))
print("\n 2..Reshaped array :\n",reshaped)

flattened = reshaped.flatten()
print("\n 3..Flattened array :\n",flattened)

# ravel (returns view,instead of copy)
raveled = reshaped.ravel()
print("\n 4..Raveled array :\n",raveled)

# transpose
transpose=reshaped.T
print("\n 5..Transposed array :\n",transpose)
```

```python
salary = np.array([300000, 56000, 78000, 23000, 65000, 10000])
salary = salary.astype(float)
salary[salary < 30000] *= 1.10
salary[(salary >= 30000) & (salary <= 60000)] *= 1.05
salary[salary > 60000] *= 1.02

print(salary)

```
